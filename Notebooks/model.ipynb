```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
DATASET_PATH = "dataset\master_alphabets_train.csv"
Test_Data = "dataset\master_alphabets_test.csv"
df = pd.read_csv(DATASET_PATH)
df.head()
df_test = pd.read_csv(Test_Data)
```

```python
FEATURE_COLS = ["flex_1", "flex_2", "flex_3", "flex_4", "flex_5"]

X = df[FEATURE_COLS]
Y = df["label"]
X_test_1 = df_test[FEATURE_COLS]
Y_test_1 = df_test["label"]

```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, Y,
    test_size=0.2,
    random_state=42,
    stratify=Y
)
```

```python
rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)
```

```python
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
y_pred = rf.predict(X_test_1)


accuracy = accuracy_score(Y_test_1, y_pred)
print("Accuracy:", accuracy)

print("\nClassification Report:")
print(classification_report(Y_test_1, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(Y_test_1, y_pred))
```
